# Reshape, stacking, squeesing, unsqueezing and view tensors3


* Squeeze - remove all the `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a tensor
* permute - Swaped the dimension of a tensor and return


```python
tensor_x = torch.arange(1., 10.)
tensor_x
```

```python
tensor_x.shape
```

```python
tensor_x.ndim
```

```python
x_reshaped = tensor_x.reshape(3,3)
x_reshaped
```

```python
x_reshaped2 = tensor_x.reshape(1,9)
x_reshaped2
```

```python
x_reshaped3 = tensor_x.reshape(9,1)
x_reshaped3
```

#### View

```python
z = tensor_x.view(1,9)
z, z.shape
```

```python
# View of the tensor share the same memory as the original input therefore if we change the view . the original will change too
```

#### Stack

```python
x_stacked = torch.stack([tensor_x,tensor_x,tensor_x,tensor_x], dim=0)
x_stacked
```

```python
x_stacked.shape
```

```python
x_stacked = torch.stack([tensor_x,tensor_x,tensor_x,tensor_x], dim=1)
x_stacked
```

```python
x_stacked.shape
```

#### Puting tensors and models into GPU

```python
# Create a tensor
tensor = torch.tensor([1, 2, 3, 4, 5], device='cpu')
tensor
```

```python
tensor.device
```

```python
# put the tensor to GPU if avaliable
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
tensor_on_gpu.device
```

```python
# Move the tensor back to CPU
tensor_to_cpu = tensor_on_gpu.to('cpu')
tensor_to_cpu
```

```python
tensor_to_cpu.device
```
